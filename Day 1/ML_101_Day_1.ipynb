# **Import necessary libraries**

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
```

# **Load the Iris dataset**

```python
iris = load_iris()
data = pd.DataFrame(iris.data, columns=iris.feature_names)
data['species'] = iris.target
```

# **Map target numbers to species names**

```python
data['species'] = data['species'].map({0: 'setosa', 1: 'versicolor', 2: 'virginica'})
```

# **4. Visualize relationships between features**

```python
# Scatter plot to visualize Sepal Length vs Sepal Width by species
plt.figure(figsize=(8, 6))

# Plotting each species separately with distinct colors
plt.scatter(data[data['species'] == 'setosa']['sepal length (cm)'],
            data[data['species'] == 'setosa']['sepal width (cm)'],
            color='red', label='Setosa')

plt.scatter(data[data['species'] == 'versicolor']['sepal length (cm)'],
            data[data['species'] == 'versicolor']['sepal width (cm)'],
            color='green', label='Versicolor')

plt.scatter(data[data['species'] == 'virginica']['sepal length (cm)'],
            data[data['species'] == 'virginica']['sepal width (cm)'],
            color='blue', label='Virginica')

# Adding labels and title
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')
plt.title('Sepal Length vs Sepal Width by Species')

# Adding a legend to show species colors
plt.legend()

# Show the plot
plt.show()
```

```python
# Scatter plot to visualize Petal Length vs Petal Width by species
plt.figure(figsize=(8, 6))

# Plotting Petal Length vs Petal Width by species
plt.scatter(data[data['species'] == 'setosa']['petal length (cm)'],
            data[data['species'] == 'setosa']['petal width (cm)'],
            color='red', label='Setosa')
plt.scatter(data[data['species'] == 'versicolor']['petal length (cm)'],
            data[data['species'] == 'versicolor']['petal width (cm)'],
            color='green', label='Versicolor')
plt.scatter(data[data['species'] == 'virginica']['petal length (cm)'],
            data[data['species'] == 'virginica']['petal width (cm)'],
            color='blue', label='Virginica')
plt.xlabel('Petal Length (cm)')
plt.ylabel('Petal Width (cm)')
plt.title('Petal Length vs Petal Width by Species')
plt.legend()

# Show the plot
plt.show()
```

```python
# Scatter plot to visualize Sepal Length vs Sepal Width by species
plt.figure(figsize=(12, 6))

# Plotting Sepal Length vs Sepal Width by species
plt.subplot(1, 2, 1)  # Create the first subplot
plt.scatter(data[data['species'] == 'setosa']['sepal length (cm)'],
            data[data['species'] == 'setosa']['sepal width (cm)'],
            color='red', label='Setosa')
plt.scatter(data[data['species'] == 'versicolor']['sepal length (cm)'],
            data[data['species'] == 'versicolor']['sepal width (cm)'],
            color='green', label='Versicolor')
plt.scatter(data[data['species'] == 'virginica']['sepal length (cm)'],
            data[data['species'] == 'virginica']['sepal width (cm)'],
            color='blue', label='Virginica')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')
plt.title('Sepal Length vs Sepal Width by Species')
plt.legend()

# Plotting Petal Length vs Petal Width by species
plt.subplot(1, 2, 2)  # Create the second subplot
plt.scatter(data[data['species'] == 'setosa']['petal length (cm)'],
            data[data['species'] == 'setosa']['petal width (cm)'],
            color='red', label='Setosa')
plt.scatter(data[data['species'] == 'versicolor']['petal length (cm)'],
            data[data['species'] == 'versicolor']['petal width (cm)'],
            color='green', label='Versicolor')
plt.scatter(data[data['species'] == 'virginica']['petal length (cm)'],
            data[data['species'] == 'virginica']['petal width (cm)'],
            color='blue', label='Virginica')
plt.xlabel('Petal Length (cm)')
plt.ylabel('Petal Width (cm)')
plt.title('Petal Length vs Petal Width by Species')
plt.legend()

# Show the plot
plt.tight_layout()  # Adjust spacing between plots
plt.show()
```
